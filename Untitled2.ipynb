```python
from tensorflow.keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train
```

```python
import numpy as np
X = np.vstack([
    X_train.reshape(-1, 784),
    X_test.reshape(-1, 784)
])
```

```python
y = np.concatenate([y_train, y_test])
```

```python
from sklearn.utils import Bunch
mnist = Bunch(
    data=X,
    target=y,
    DESCR="MNIST dataset"
)

print(mnist.keys())
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
sgd_clr=SGDClassifier(random_state=42)
sgd_clr.fit(x_train , y_train_5)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
srk=StratifiedKFold(n_splits=3 , random_state=42, shuffle=True)
for train_x , test_x in srk.split(x_train , y_train_5):
    clone=clone(sgd_clr)
    x_train_fold=x_train[train_x]
    y_train_fold=y_train_5[train_x]
    X_test_fold=x_train[test_x]
    y_test_fold=y_train_5[test_x]

    clone.fit(x_train_fold , y_train_fold)
    y_pred = clone.predict(X_test_fold)  
    n_correct = (y_pred == y_test_fold).sum()
    print(n_correct / len(y_test_fold))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clr , x_train , y_train_5 ,cv=3, scoring='accuracy')
```

```python
from sklearn.base import BaseEstimator 
class pipeline_est(BaseEstimator):
    def fit(self , X, y=None):
        return self

    def predict(self, X):
        return np.zeros((len(X) , 1 ), dtype=bool)

    
```

```python
never_s=pipeline_est()
cross_val_score(never_s , x_train , y_train_5, cv=3 , scoring='accuracy')

```
